```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Read in scores csv
scores_data = pd.read_csv("../data/naep.csv")
# Filter data to include 8th Grade, years 2003 onwards and exclude the Department of Defense Education Activity
# Note: Even years scores are the average of the in-btwn odd years. E.g. 2014_score = average(2013_score,2015_score)
scores = scores_data.loc[(scores_data["TEST_YEAR"]==8)&(scores_data["YEAR"]>=2003)&(scores_data["STATE"]!= "DoDEA")].reset_index(drop=True)
# Check for any other irrelevant states/locations
scores["STATE"].value_counts()
# Rename column headers
scores = scores.rename(columns={"YEAR":"Year","STATE":"State","AVG_SCORE":"AverageScore","TEST_SUBJECT":"Subject","TEST_YEAR":"Grade"})
# Preview cleaned df
scores.shape
scores.head()
```

```python
# Read in funding by state csv
funding_data = pd.read_csv("../data/states_funding.csv")
# Filter file to include years 2003 onwards and reset index
funding = funding_data.loc[funding_data["YEAR"]>=2003].reset_index(drop=True)
# Get rid of expenditure columns as not needed
funding = funding.drop(columns=["INSTRUCTION_EXPENDITURE","SUPPORT_SERVICES_EXPENDITURE","OTHER_EXPENDITURE","CAPITAL_OUTLAY_EXPENDITURE"])
# Rename column headers to look better
funding = funding.rename(columns={"STATE":"State","YEAR":"Year","ENROLL":"StudentsEnrolled","TOTAL_REVENUE":"TotalFunding","FEDERAL_REVENUE":"FederalFunding","STATE_REVENUE":"StateFunding","LOCAL_REVENUE":"LocalFunding","TOTAL_EXPENDITURE":"TotalExpenditure"})
# Preview cleaned df
# funding["Year"].value_counts()
funding.head()
```

```python
# Merge Scores and Funding dataframes
scores_and_funding = pd.merge(scores, funding, on=["Year","State"],how="left")\
# Calculate $ received per student enrolled and add new column
scores_and_funding["DollarsPerStudent"] = scores_and_funding["TotalFunding"]/scores_and_funding["StudentsEnrolled"]
# Preview combined file
scores_and_funding.head()
```

```python
# Check for inconsistent characters
scores_and_funding["AverageScore"].value_counts()
# Replace non-numerical characters with nan or 0
scores_and_funding["AverageScore"] = scores_and_funding["AverageScore"].replace({"—":0,"‡":np.nan})
scores_and_funding["StateFunding"] = scores_and_funding["StateFunding"].replace({"(X)":0})
```

```python
# Convert Average Score column to float type
scores_and_funding["AverageScore"] = scores_and_funding["AverageScore"].astype(float)
scores_and_funding["StateFunding"] = scores_and_funding["StateFunding"].astype(float)
# Check types are correct for calculations
scores_and_funding.dtypes
```

```python
# Read in State Abbreviations CSV
state_Abbrs = pd.read_csv("../data/stateAbbrs.csv")
# Drop unnecessary columns
state_Abbrs = state_Abbrs[["State","Abbreviation"]]
# Merge Abbrs to scores and funding dataset
scores_and_funding = pd.merge(scores_and_funding,state_Abbrs,on="State",how="left")
scores_and_funding.head()
```

## Begin training model

```python
# Select relevant X and y variables to be passed into model
X = final_education_train_df.drop(["Year","Grade","AverageScore","Abbreviation"],axis=1)
y = final_education_train_df["AverageScore"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
X.head()
```

```python
# One-hot encoding to convert categorical data into numerical data
data = X.copy()
X = pd.get_dummies(data, columns=["State","Subject"])
X.head()
```

```python
# Split the data to have a training data batch through which to train the model, and a test data batch for 
# subsequent model testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train.head()
```

```python
X_train.shape
```

```python
# Fit the X variables to the scaler
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
# Scale the X variables so no variable overpowers another
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Plot the results to view residuals
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train, c="#1380EC", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test, c="#FF69B4", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test.min(), xmax=y_test.max())
plt.title("Residual Plot")
plt.savefig("../images/residuals")
plt.show()
```

```python
# Run the scaled data and test data through the model. Check reasonableness with MSE and R2 scores
from sklearn.metrics import mean_squared_error

predictions = model.predict(X_test_scaled)
MSE = mean_squared_error(y_test, predictions)
r2 = model.score(X_test_scaled, y_test)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
# Compare 1st item in trained model prediction to actual 1st item in test data batch
print(predictions[0])
print(y_test[0])
```

```python
# Check model Accuracy
print('Test Acc: %.3f' % model.score(X_test_scaled, y_test))
```

```python
# Check coefficients for how strongly each variable impacts the average score
model.coef_
```
